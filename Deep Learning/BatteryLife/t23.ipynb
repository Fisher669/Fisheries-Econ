```python
import torch
import torch.nn as nn
from einops import rearrange

def to_3d(x):
    return rearrange(x, 'b c h w -> b (h w) c')
def to_4d(x, h, w):
    return rearrange(x, 'b (h w) c -> b c h w', h=h, w=w)

class SELayer(nn.Module):
    def __init__(self, channel, reduction=16):
        super(SELayer, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channel, channel // reduction, bias=False),
            nn.ReLU(inplace=True),
            nn.Linear(channel // reduction, channel, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.size()
        y = self.avg_pool(x).view(b, c)
        y = self.fc(y).view(b, c, 1, 1)
        return x * y.expand_as(x)

class SEbox(nn.Module):
    def __init__(self, in_channel, out_channel, kernel_size, stride, padding):
        super(SEbox, self).__init__()
        self.conv = nn.Conv2d(in_channel, out_channel, kernel_size, stride, padding)
        self.bn = nn.BatchNorm2d(out_channel)
        self.relu = nn.LeakyReLU(inplace=True)
        self.pool_out = nn.AdaptiveMaxPool2d(output_size=(8,19))
        self.se = SELayer(out_channel)
        self.cbam = CBAM(out_channel)
        self.dropout = nn.Dropout(p=0.5)

    def forward(self, x):
        x = self.conv(x)
        x = self.bn(x)
        x = self.relu(x)
        x = self.pool_out(x)
        x = x + self.se(x)
        x = self.dropout(x)
        x = self.cbam(x)
        return x
    
class cnn_box(nn.Module):
    def __init__(self, in_channel, out_channel, kernel_size, stride, padding):
        super(cnn_box, self).__init__()
        self.conv = nn.Conv2d(in_channel, out_channel, kernel_size, stride, padding)
        self.bn = nn.BatchNorm2d(out_channel)
        self.relu = nn.LeakyReLU(inplace=True)
        self.dropout = nn.Dropout(p=0.5)

    def forward(self, x):
        x = self.conv(x)
        x = self.bn(x)
        x = self.relu(x)
        x = self.dropout(x)
        return x
    
class CBAM(nn.Module):
    def __init__(self, in_channel, reduction=16, kernel_size=7):
        super(CBAM, self).__init__()
        self.max_pool = nn.AdaptiveMaxPool2d(output_size=1)
        self.avg_pool = nn.AdaptiveAvgPool2d(output_size=1)
        self.mlp = nn.Sequential(
            nn.Linear(in_features=in_channel, out_features=in_channel // reduction, bias=False),
            nn.ReLU(),
            nn.Linear(in_features=in_channel // reduction, out_features=in_channel, bias=False)
        )
        self.sigmoid = nn.Sigmoid()
        self.conv = nn.Conv2d(in_channels=2, out_channels=1, kernel_size=kernel_size, stride=1, padding=kernel_size // 2, bias=False)

    def forward(self, x):
        maxout = self.max_pool(x)
        avgout = self.avg_pool(x)
        channel_out = self.sigmoid(self.mlp(maxout.view(maxout.size(0), -1)) + self.mlp(avgout.view(avgout.size(0), -1)))
        channel_out = channel_out.view(x.size(0), x.size(1), 1, 1) * x
        max_out = torch.max(channel_out, dim=1, keepdim=True)[0]
        mean_out = torch.mean(channel_out, dim=1, keepdim=True)
        out = self.sigmoid(self.conv(torch.cat((max_out, mean_out), dim=1))) * channel_out
        return out

class InceptionModule(torch.nn.Module):
    def __init__(self, in_channels, out_channels):
        super(InceptionModule, self).__init__()
        
        self.inceptions = torch.nn.ModuleList([
            torch.nn.Sequential(
                torch.nn.Conv2d(in_channels, out_channels, kernel_size=(1, 1), stride=1),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.LeakyReLU(inplace=True)
            ),
            torch.nn.Sequential(
                torch.nn.Conv2d(in_channels, out_channels, kernel_size=(1, 1), stride=1),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.LeakyReLU(inplace=True),
                torch.nn.Conv2d(out_channels, out_channels, kernel_size=(3, 3), stride=1, padding=1),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.LeakyReLU(inplace=True)
            ),
            torch.nn.Sequential(
                torch.nn.Conv2d(in_channels, out_channels, kernel_size=(1, 1), stride=1),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.LeakyReLU(inplace=True),
                torch.nn.Conv2d(out_channels, out_channels, kernel_size=(5, 5), stride=1, padding=2),
                torch.nn.BatchNorm2d(out_channels),
                torch.nn.LeakyReLU(inplace=True)
            ),
            torch.nn.Sequential(
                torch.nn.MaxPool2d(kernel_size=3, stride=1, padding=1),
                torch.nn.Conv2d(in_channels, out_channels, kernel_size=(1, 1), stride=1),
                torch.nn.BatchNorm2d(out_channels),
            )
        ])

    def forward(self, x):
        # 将输入传递给所有Inception模块，并将输出连接起来
        outputs = [inception(x) for inception in self.inceptions]
        return torch.cat(outputs, dim=1)# 这里按通道维度连接输出

class Go_lstm(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(Go_lstm, self).__init__()
        self.go= InceptionModule(in_channels, out_channels)
        # self.fc = nn.Linear(in_channels*4, out_channels)
        self.dropout = nn.Dropout(p=0.5)
        self.lstm = nn.LSTM(input_size=out_channels*4, hidden_size=out_channels, num_layers=1, batch_first=True)    
        # self.deduction = nn.AdaptiveMaxPool2d(output_size=(8,19))
           
    def forward(self, x):
        # print(x.shape)
        x = self.go(x)          
        x = to_3d(x)
        x, _ = self.lstm(x)
        # print(x.shape)
        x = to_4d(x, 8, 19)
        # print(x.shape)
        return x

# 使用示例
# model = InceptionModule(in_channels=32, out_channels=64)
# output = model(input_tensor)  # input_tensor 是你的输入数据
```

```python

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.layers1 = nn.Sequential(
            nn.Conv2d(1, 8, kernel_size=(50, 50), stride=(5, 40)),
            nn.BatchNorm2d(8),
            nn.LeakyReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=1),
            nn.Dropout(p=0.2),
            
            torch.nn.Conv2d(8, 16, kernel_size=(2, 3), stride=1),
            torch.nn.BatchNorm2d(16),
            torch.nn.LeakyReLU(inplace=True),
            torch.nn.Dropout(p=0.2),
            
            torch.nn.Conv2d(16, 32, kernel_size=(2, 3), stride=1),
            torch.nn.BatchNorm2d(32),
            torch.nn.LeakyReLU(inplace=True),            
            
        )
        self.golstm1 = Go_lstm(32,32)
        self.dropout = nn.Dropout(p=0.2)
        self.golstm2 = Go_lstm(32,32)
        self.golstm3 = Go_lstm(32,32)
        self.fc1 = nn.Linear(4864, 1)

    def forward(self, x):
        try:
            # 网络结构部分
            x = self.layers1(x)
            x = self.golstm1(x)
            x = self.dropout(x)
            x = self.golstm2(x)
            x = self.dropout(x)
            x = self.golstm3(x)
            x = self.dropout(x)
            # print(f'after golstm: {x.shape}')
            x = x.reshape(x.size(0), -1)
            # print(f'after reshape: {x.shape}')
            return torch.sigmoid(self.fc1(x))
        except RuntimeError as e:
            print(f"运行时错误：{e}")
            return None
        except Exception as e:
            print(f"发生意外错误：{e}")
            raise  # 重新抛出异常，以便进一步处理
```

```python

net = Net()     # define the network
print(net)      # net architecture
# summary(net,(1,100,1000))
```

```python
import pandas as pd
import torch.optim as optim
from tqdm import tqdm  # 加入进度条
import datetime
import os
import torch

# Data loader for Pytorch
input_train = torch.FloatTensor(train_Qd)
input_train = torch.unsqueeze(input_train, 1)
train_labels = torch.FloatTensor(train_cycle_life)

input_val = torch.FloatTensor(train_Qds)
input_val = torch.unsqueeze(input_val, 1)
val_labels = torch.FloatTensor(train_cycle_lifes)

input_test_a = torch.FloatTensor(test_Qd_a)
input_test_a = torch.unsqueeze(input_test_a, 1)
test_labels_a = torch.FloatTensor(test_cycle_life_a)

input_test_b = torch.FloatTensor(test_Qd_b)
input_test_b = torch.unsqueeze(input_test_b, 1)
test_labels_b = torch.FloatTensor(test_cycle_life_b)

seed = 17
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.cuda.manual_seed_all(seed)  # if you are using multi-GPU.
np.random.seed(seed)  # Numpy module.
random.seed(seed)  # Python random module.
torch.manual_seed(seed)
torch.backends.cudnn.benchmark = False
torch.backends.cudnn.deterministic = True
torch.manual_seed(seed)    # reproducible    

optimizer_adam = torch.optim.Adam(net.parameters(), lr=0.0001)
optimizer_sgd = torch.optim.SGD(net.parameters(), lr=0.0001,momentum=0.9)
loss_func = torch.nn.MSELoss() 

scheduler_adam = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_adam, T_max=100)
# scheduler_sgd = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_sgd, T_max=100)
scheduler_sgd = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer_sgd, mode='min', factor=0.1, patience=25)
# scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=10)
# 创建历史记录DataFrame
history = pd.DataFrame(columns=['index', 'train_loss', 'val_loss'])
val_losses = []
max_models_to_keep = 10
saved_models = []

log_file_path = 'train_log.txt'  # 请替换为实际的日志文件路径

# 初始化日志文件，并且第一行记录时间跟训练参数
with open(log_file_path, 'w') as log_file:
    log_file.write('Time: {}\n'.format(datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')))
    # log_file.write('Training parameters: {}\n'.format(str(vars(args))))

# 使用tqdm添加进度条
for epoch in range(0, 1000, 200):
    with tqdm(range(epoch, epoch + 200), desc=f"Training Progress {epoch}-{epoch + 200}") as progress_bar:
        for t in progress_bar:
            net.train()
            train_prediction = net(input_train)
            train_loss = loss_func(train_prediction, train_labels)
            if t>200:
                optimizer = optimizer_adam
                scheduler = scheduler_adam
                optimizer.zero_grad()
                train_loss.backward()
                optimizer.step()
                scheduler.step()
            else:
                optimizer = optimizer_sgd
                scheduler = scheduler_sgd
                optimizer.zero_grad()
                train_loss.backward()
                optimizer.step()
                scheduler.step(val_loss)

            net.eval()
            with torch.no_grad():
                val_prediction = net(input_val)
                val_loss = loss_func(val_prediction, val_labels)

            # 更新进度条描述信息
            progress_bar.set_postfix(train_loss=train_loss.item(), val_loss=val_loss.item())

            # 打开文件并写入内容
            with open(log_file_path, 'a') as log_file:  # 'a'模式表示追加写入
                log_file.write('Step = %d, train_loss: %s, val_loss: %s\n' % (t, train_loss.item(), val_loss.item()))

            val_losses.append(val_loss.item())
            
            # 添加当前模型及其验证损失
            saved_models.append((val_loss.item(), f'Target_model/net_parameters{t}.pkl'))

            # 保存模型
            torch.save({
                'epoch': t,
                'model_state_dict': net.state_dict(),
                'optimizer_state_dict': optimizer.state_dict()}, f'Target_model/net_parameters{t}.pkl')

            # 保持模型数量不超过max_models_to_keep
            if len(saved_models) > max_models_to_keep:
                # 找到验证损失最大的一组模型并删除
                saved_models.sort(key=lambda x: x[0])  # 排序，根据损失
                os.remove(saved_models.pop()[1])  # 删除损失最大的模型

            # 加入训练历史记录
            new_entry = pd.DataFrame({'index': [t], 'train_loss': [train_loss.item()], 'val_loss': [val_loss.item()]})
            history = pd.concat([history, new_entry], ignore_index=True)

'''find the best model'''
best_index = val_losses.index(np.min(val_losses))
print(f'Best model at index: {best_index}')
print(f'Best model val_loss: {val_losses[best_index]}')

'''copy the best model from model file to the model_test file'''
shutil.copyfile(f'Target_model/net_parameters{best_index}.pkl', 'Best_target_model/net_parameters.pkl')

'''Reload the best model''' 
model = Net()
checkpoint = torch.load('Best_target_model/net_parameters.pkl')
model.load_state_dict(checkpoint['model_state_dict'])
optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
```

```python

'''Prediction using the best model'''
predict_labels_a=[]
predict_labels_b=[]
predict_labels_train=[]
predict_labels_val = []
model.eval()
with torch.no_grad():
    predict_labels_a.append(model(input_test_a).data.numpy())
    predict_labels_b.append(model(input_test_b).data.numpy())
    
    predict_labels_train.append(model(input_train).data.numpy())   
    predict_labels_val.append(model(input_val).data.numpy())      


actual_label_arr_train=np.round(np.asarray(train_labels*max_label).reshape(-1,1))  
predict_label_arr_train=np.round(np.asarray(predict_labels_train).reshape(-1,1)*max_label)


actual_label_arr_val=np.round(np.asarray(val_labels*max_label).reshape(-1,1))   
predict_label_arr_val=np.round(np.asarray(predict_labels_val).reshape(-1,1)*max_label)


actual_label_arr_a=np.round(np.asarray(test_cycle_life_a*max_label).reshape(-1,1))   
predict_label_arr_a=np.round(np.asarray(predict_labels_a).reshape(-1,1)*max_label)


actual_label_arr_b=np.round(np.asarray(test_cycle_life_b * max_label).reshape(-1,1))   
predict_label_arr_b=np.round(np.asarray(predict_labels_b).reshape(-1,1)*max_label)

```
